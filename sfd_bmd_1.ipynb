```python
import numpy as np
import matplotlib.pyplot as plt

# Define segment lengths
x1 = 3  # ft
x2 = 6
x3 = 3
L = x1 + x2 + x3  # Total length

# Generate points along the beam
a = np.linspace(0, L,1000)
m = np.zeros_like(a)  # Initialize moment array

# Define moment equations for different sections
m[np.logical_and(0 < a, a < x1)] = 1900 * a[np.logical_and(0 < a, a < x1)]
m[np.logical_and(x1 < a, a < x1 + x2)] = 1900 * a[np.logical_and(x1 < a, a <x1 + x2)] -\
    2000 * (a[np.logical_and(x1 < a, a < x1 + x2)] - x1)
m[np.logical_and(x1 + x2 < a, a < L)] = 1900 * a[np.logical_and(x1 + x2 < a, a < L)] -\
    2000 * (a[np.logical_and(x1 + x2 < a, a < L)] - x1) - 4800

# Plotting the moment diagram
plt.plot(a,m)
plt.xlabel("Length along the beam (ft)")
plt.ylabel("Bending Moment (lb-ft)")
plt.title("Bending Moment Diagram")
plt.axhline(0, color='black', linewidth=0.5)  # X-axis reference line
plt.legend()
plt.grid()
plt.show()




```
